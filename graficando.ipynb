```python
import numpy as np
from matplotlib.colors import ListedColormap
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

# Seleccionar los datos de entrenamiento
X = datos2022_cantidad[['num_salas', 'num_docentes']].values
y = datos2022_cantidad['acreditacion_institucional'].values

# Crear el imputador
imp = SimpleImputer(strategy='mean')

# Ajustar el imputador a los datos y luego transformar los datos
X_imputed = imp.fit_transform(X)

# Normalizar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

# Crear y entrenar el modelo KNN
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_scaled, y)

# Crear colormaps
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00'])

# Crear una cuadrícula para evaluar el modelo
x_min, x_max = X_scaled[:, 0].min() - 1, X_scaled[:, 0].max() + 1
y_min, y_max = X_scaled[:, 1].min() - 1, X_scaled[:, 1].max() + 1
h = .02
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])

# Mostrar el resultado en un gráfico de colores
Z = Z.reshape(xx.shape)
plt.figure(figsize=(10,5))
plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

# También plotear los puntos de entrenamiento
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y, cmap=cmap_bold)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.title("Clasificación de universidades 2022 (K = 3)")
plt.xlabel('Número de salas normalizadas')
plt.ylabel('Número de docentes normalizadas')
plt.show()
```
