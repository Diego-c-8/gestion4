```python
##limpiar datos de años de acreditacion de objet a int
institucion['anos_acreditacion'] = pd.to_numeric(institucion['anos_acreditacion'], errors='coerce').fillna(0).astype(int)
institucion['acreditacion_institucional'] = institucion['acreditacion_institucional'].apply(lambda x: 1 if x == 'Acreditada' else 0)
institucion['gestion_institucional'] = institucion['gestion_institucional'].apply(lambda x: 1 if x == 'Si' else 0)
institucion['docencia_pregrado'] = institucion['docencia_pregrado'].apply(lambda x: 1 if x == 'Si' else 0)
institucion['investigacion'] = institucion['investigacion'].apply(lambda x: 1 if x == 'Si' else 0)
institucion['vinculacion_medio'] = institucion['vinculacion_medio'].apply(lambda x: 1 if x == 'Si' else 0)
institucion['docencia_postgrado'] = institucion['docencia_postgrado'].apply(lambda x: 1 if x == 'Si' else 0)
institucion['vigencia_desde'] = pd.to_datetime(institucion['vigencia_desde'], errors='coerce')
institucion['vigencia_hasta'] = pd.to_datetime(institucion['vigencia_hasta'], errors='coerce')
# Verificar el resultad
```

```python
import pandas as pd
import sqlite3

# Crear una conexión a la base de datos SQLite
conexion = sqlite3.connect('nombre_basedatos.db')

# DataFrames que quieres insertar
dataframes = {
    'tabla1': institucion,
    'tabla2': inmuebles,
    'tabla3': laboratorios_talleres,
    'tabla4': bibliotecas,
    'tabla5': docentes
}

# Iterar sobre los DataFrames y guardarlos como tablas en la base de datos
for nombre_tabla, dataframe in dataframes.items():
    dataframe.to_sql(nombre_tabla, conexion, if_exists='replace', index=False)

# Cerrar la conexión a la base de datos
conexion.close()
```
